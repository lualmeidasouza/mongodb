```python
#Operações CRUD no MongoDB usando Python

#O que significa Crud? C-Create, R-Retrieve, U-Update,  D-Delete no MongoDB 

#It was a milestone in my life from developer: SQL to NoSQL. #awesome

"""
-MongoDB é um banco de dados NoSQL orientado a documentos.
-Um dos bancos de dados mais utilizados na indústria devido à sua flexibilidade e facilidade de uso. 
-A arqutetura do MongoDB consiste em coleções e documentos.
-O MongoDB armazena dados no formato de documentos JSON, ou sea, ele armazena dados no par de valor-chave. 
-Um grupo de documentos JSON pode ser denominado como uma coleção. 
-Ao contrário das tabelas SQL, os documentos MongoDB não têm nenhum esquema fixo. 
-Com o uso de esquema dinâmico, podemos facilmente fazer alterações no aplicativo, sem interrupções.

INGREDIENTES:
Você vai precisar de:
    - SO Windows 10 ou superior... Ou Linux! :D
    - Sua IDE preferida (eu uso Spyder, mas você pode escoler PyCharm... Ou outra de sua preferência)
    - MongoDB Compass (baixe aqui: https://www.mongodb.com/try/download/compass)
    - A partir do prompt do DOS, instale a bilioteca pymongo para interagir com o MongoDB a partir do Python. No DOS, digite: pip install pymongo

"""
import pymongo

#Biblioteca para manipulação de dataframes
import pandas as pd

#String de conexão. Para se conectar ao servidor MongoDB, usaremos o método MongoClient.
client = pymongo.MongoClient("mongodb://myUser:myPassword@localhost:27017/?readPreference=primary&authSource=admin&appname=MongoDB%20Compass&ssl=false")

#Existem três bancos de dados por padrão, que são admin, config e local. Vamos criar o nosso banco chamado "estudoMongoDB"
db = client["estudoMongoDB"]

#Criamos uma collection chamada 
alunos = db["alunos"]

#O comando insert como o próprio nome já diz insere um document na collection referenciada. Então, vamos popular!!!

db.alunos.insert({
  "nome": "Maria",
  "dataNascimento": "17/05/1981",
  "notas": {
    "matematica": 9.5,
    "fisica": 8,
    "portugues": 10,
    "geografia": 8,
  },
  "peso": 71,
  "altura": 1.66,
  "sexo": "F",
  "pais": "Brasil",
  "estado": "SP"
})

db.alunos.insert({
  "nome": "João",
  "dataNascimento": "17/05/1981",
  "notas": {
    "artes": 9.5,
    "musica": 8,
    "ingles": 10,
    "ciencias": 8,
  },
  "peso": 71,
  "altura": 1.66,
  "sexo": "F",
  "pais": "Brasil",
  "estado": "SP"
})

db.alunos.insert({
  "nome": "Paulo",
  "dataNascimento": "17/05/1981",
  "notas": {
    "matematica": 9,
    "fisica": 8,
    "portugues": 7,
    "geografia": 5,
  },
  "peso": 71,
  "altura": 1.66,
  "sexo": "F",
  "pais": "Brasil",
  "estado": "SP"
})


db.alunos.insert({
  "nome": "Miguel",
  "dataNascimento": "17/05/1981",
  "notas": {
    "matematica": 5,
    "fisica": 6,
    "portugues": 9,
    "geografia": 8,
  },
  "peso": 71,
  "altura": 1.66,
  "sexo": "F",
  "pais": "Brasil",
  "estado": "SP"
})

db.alunos.insert({
  "nome": "Anderson",
  "dataNascimento": "17/05/1981",
  "notas": {
    "matematica": 5,
    "fisica": 5,
    "portugues": 5,
    "geografia": 5,
  },
  "peso": 71,
  "altura": 1.66,
  "sexo": "F",
  "pais": "Brasil",
  "estado": "SP"
})

db.alunos.insert({
  "nome": "Katiany",
  "dataNascimento": "17/05/1970",
  "notas": {
    "matematica": 9,
    "fisica": 9,
    "portugues": 9,
    "geografia": 8,
  },
  "peso": 71,
  "altura": 1.50,
  "sexo": "F",
  "pais": "Brasil",
  "estado": "SP"
})

db.alunos.insert({
  "nome": "Francieli",
  "dataNascimento": "17/05/1981",
  "notas": {
    "matematica": 9.5,
    "fisica": 8,
    "portugues": 10,
    "geografia": 8,
  },
  "peso": 71,
  "altura": 1.66,
  "sexo": "F",
  "pais": "Brasil",
  "estado": "SP"
})

#consultando
df = pd.DataFrame(list(db.alunos.find({ "nome" : "Maria", "dataNascimento" : "17/05/1981", "notas" : { "matematica" : 9.5, "fisica" : 8, "portugues" : 10, "geografia" : 8 }, "peso" : 71, "altura" : 1.66 })))
print(df)

#Joga resultado da query num dataframe
df = pd.DataFrame(list(db.alunos.find({ "nome" : "Luciana Almeida da Silveira", "dataNascimento" : "17/05/1981", "notas" : { "matematica" : 9.5, "fisica" : 8, "portugues" : 10, "geografia" : 8 }, "peso" : 71, "altura" : 1.66 })))
#print(df)

#Para indentá-lo facilitando uma melhor leitura 
df = pd.DataFrame(list(db.alunos.find({  
  "nome": "Carlos",
  "dataNascimento": "17/05/1971",
  "notas": {
    "artes": 9.5,
    "musica": 8,
    "ingles": 10,
    "ciencias": 8,
  },
  "peso": 71,
  "altura": 1.66,
  "sexo": "F",
  "pais": "Brasil",
  "estado": "SP",
  "e-mail": "carlos@ig.com.br"
})))


#print(df)

"""
O projection retorna somente os campos em que setamos como 1 ou true, 
não é preciso colocar todos os campos do document no objeto como true ou false, 
se você não irá precisar do campo _id então você deve setá-lo como 0 ou false, 
pois por padrão ele é tido como true, é importante a remoção do campo _id quando
não for usá-lo até mesmo por uma questão de segurança."""

# retorna somente os campos nome e data de nascimento
df = pd.DataFrame(list(db.alunos.find({"nome": "Luciana Silveira"}, {"_id":0, "nome":1, "dataNascimento":1})))
#print(df)

#Existem alguns tipos de chaves que usamos como uma espécie de filtos para o find, como por exemplo:
"""
gt (greater than) maior que
gte (greater than or equals) maior ou igual a
lt (less than) menor que
lte (less than or equals) menor ou igual a
"""

# retorna todos os alunos que possuiram notas iguais ou superiores a 7 em matemática
# observe que a matematica é uma propriedade do array notas que está contido dentro da collection alunos
df = pd.DataFrame(list(db.alunos.find({"notas.matematica": {"$lt":7}})))

#Regex: Você também pode usar expressões regulares nos comandos
# retorna todos os alunos que possuem a string "any" em algum lugar do nome
df = pd.DataFrame(list(db.alunos.find({"nome": {"$regex": "any"}})))

# retorna todos os alunos que começam com a letra "A" maiúsculo
df = pd.DataFrame(list(db.alunos.find({"nome": {"$regex": "^A"}})))

# retorna todos os alunos que termina com a letra "y"
df = pd.DataFrame(list(db.alunos.find({"nome": {"$regex": "y$"}})))
#print(df)

"""
operadores lógicos e / ou
Os operadores lógicos são representados pela sintaxe $and / $or e requerem um array, 
dentro desse array são passadas todas as condições para serem analisadas. 
O operador $and retorna o documento buscado, somente se todas as queries passadas forem verdadeiras. 
E o operador $or retorna o documento se pelo menos uma query for verdadeira.
"""

#Ex: db.alunos.find({ $or: [ {condição 1, condição 2} ] }).
df = pd.DataFrame(list(db.alunos.find({"$or":[ { "nome": "Carlos", "e-mail": {"$exists":"true"}}]})))
#print(df)

"""
operadores all / in - Esses operadores existem para trabalharmos com arrays.
O operador $all retorna o documento se todos os itens procurados estiverem contidos na query.
E o operador $in varre o array mostrando os documents que contenham algum dos itens especificados.
"""

# retorna os documents que possuem todos os itens no array passado
df = pd.DataFrame(list(db.alunos.find({"notas.matematica": {"$all": [9, 10]}})))
# print(df)


# retorna os documents que possuem pelo menos un dos itens no array passado
df = pd.DataFrame(list(db.alunos.find({"notas.matematica": {"$in": [5, 6]}})))
#print(df)


"""
UPDATE COM REPLACE - Existem algumas maneiras de fazer Update diretamente no document,
uma delas é o comando update sem nenhuma outra feature, que atualiza o document 
para exatamente o objeto passado por parâmetro.

Como você pode ver a função update recebe dois objetos como parâmetro, 
o primeiro é a query para busca, e o segundo objeto é o replace, ou seja, 
são os dados que substituirão aquele document.

OBS: Quando se usa o update da maneira acima, os elementos anteriores da tabela 
são descartados com excessão do _id.

Por isso, é importante colocar o valor de replace correto.
"""


df = pd.DataFrame(list(db.alunos.update({"nome": 'Maria Maria'},{"nome": 'Carlos Miguel', "dataNascimento": "17/05/1971",
  "notas": {
    "artes": 9.5,
    "musica": 8,
    "ingles": 10,
    "ciencias": 8,
  },
  "peso": 71,
  "altura": 1.66,
  "sexo": "F",
  "pais": "Brasil",
  "estado": "SP",
  "e-mail": "miguelcarlos@ig.com.br"})))
#print(df)


"""
UPDATE COM SET (recomendado): o operador set é usado quando queremos evitar que a substituição inteira do document aconteça, 
ou seja, quando queremos acrescentar e / ou alterar somente alguns campos no document, 
sem apagar todos os outros.
"""
df = pd.DataFrame(list(db.alunos.update({"nome": 'Anderson'},{"$set": {"peso":85}})))
#print(df)

"""
UPSERT (update + insert): Cria um novo document, caso não exista nenhum com as características passadas
"""
df = pd.DataFrame(list(db.alunos.update({"nome": "Anderson"}, {"$set": {"peso": 90}}, upsert=True)))
#print(df)

"""
INC: Incrementa os valores ao elemento.
"""
df = pd.DataFrame(list(db.alunos.update({"nome": 'Anderson' },{"$inc": {"peso":1} })))
#print(df)

"""
UNSET: Remove um atributo de um ou mais documentos, conforme a condição especificada.
"""
df = pd.DataFrame(list(db.alunos.update({"nome": "Katiany"},{'$unset': {"sexo": "F"}}, multi=True)))                                                           
#print(df)

#UPDATE EM ARRAYS: Existem as seguintes maneiras de fazermos updates em um array, são elas:
#PUSH: acrescenta o item ao final do array.
df = pd.DataFrame(list(db.alunos.update({"nome": 'Luciana Silveira'},{"$push": {"notas.teatro": {'artes': 10}}})))
#print(df)


#POP: Remove itens por índice (posição) do array.
#df = pd.DataFrame(list(db.alunos.update_one({"_id": '6197a15824930a7509fc9bc7'},{"$pop": {"notas.portugues": 10}})))
#print(df)


# pushAll: Acrescenta mais de um elemento ao array
#df = pd.DataFrame(list(db.alunos.update({"nome": "Carlos Miguel"}, {"$pushAll": {"notas": ["geografia", "historia"]}})))
#print(df)

#pull: Remove item por valor.
#df = pd.DataFrame(list(db.alunos.update({"nome": "Francieli" },{"$pull": {"notas": 2 } })))
#print(df)

#pullAll: Remove mais de um item no array, possui a mesma sintaxe do pushAll.
#df = pd.DataFrame(list(db.alunos.update({"nome": 'Anderson'  },{"$pullAll": {"notas": ["matematica", "ingles"]}})))
#print(df)

#addToSet: Acrescenta o elemento ao array somente se ele não existir.
#db.array.update({"id": 0  },{"$addToSet": {"a": 5 } })

#Multi update: quando precisamos fazer o update em mais de um elemento, precisamos setar o multi update.
#db.alunos.update({}, {$set: {"nome": "Anderson"}}, {multi="True"})

#Remove: possui a mesma sintaxe do comando find.
#Atenção: Em algumas versões do MongoDB se for passado o objeto vazio como parâmetro, a collection inteira é removida
#db.alunos.remove({"nome":  'NoNoNo' })

#Uma outra maneira de remover a collection inteira é com o comando drop.
#db.alunos.drop()

#Este resumo está disponível aqui: https://github.com/lualmeidasouza
```
